```python
from langchain_community.vectorstores import Chroma
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.document_loaders import DataFrameLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
from langchain_huggingface import HuggingFaceEmbeddings

model = "BAAI/bge-base-en-v1.5"

embedder = HuggingFaceEmbeddings(
    model_name=model,
    model_kwargs={"device": "cuda"}
)

sentences = [
    "The weather is lovely today.",
    "It's so sunny outside!",
    "He drove to the stadium."
]

embs = embedder.embed_documents(sentences)
print(len(embs), len(embs[0]))   # (3, 768)
```

```python
import pandas as pd

df = pd.read_pickle("/content/processed_product.pkl")
df
```

# Split Data

```python
column_name = 'combined_text'

df_loader = DataFrameLoader(df, page_content_column=column_name)
df_document = df_loader.load()

CHUNK_SIZE = 1500
CHUNK_OVERLAP = 100

splitter = RecursiveCharacterTextSplitter(
        chunk_size=CHUNK_SIZE,
        chunk_overlap=CHUNK_OVERLAP,
        length_function=len,
        add_start_index=True,
    )

chunked_docs = splitter.split_documents(df_document)
```

```python
chunked_docs[0]
```

# Chroma DB

```python
%%time

# load embeddings
embeddings = HuggingFaceEmbeddings(model_name="BAAI/bge-base-en-v1.5",
                                   model_kwargs={"device": "cuda"},
                                   encode_kwargs={'batch_size': 2048})

# create vector db
print("vector db creation is started...")

chromadb_index = Chroma.from_documents(chunked_docs, embeddings, persist_directory=f'chroma_db', collection_metadata={"hnsw:space": "cosine"})

print("vector db is created successfully...")
```

```python
from typing import Any


query = "what are the top 5 most rated guitars?"

result = chromadb_index.similarity_search_with_score(query=query, k = 10)

for i, d in enumerate[Any](result):
    print(f"---- DOCUMENT {i} ----")
    print(d[0].page_content)
    print("METADATA:", d[0].metadata, "\n")
```
