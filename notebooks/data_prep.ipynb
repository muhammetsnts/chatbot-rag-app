```python
import ast
import json
import pandas as pd

def _parse_details(x):
    
    if pd.isna(x):
        return {}
    if isinstance(x, str):
        s = x.strip()
        if not s or s in ("[]", "None", "null"):
            return {}
        try:
            v = ast.literal_eval(s)
        except Exception:
            try:
                v = json.loads(s)
            except Exception:
                return {}
        if isinstance(v, dict):
            return v
        if isinstance(v, list) and len(v) > 0 and isinstance(v[0], dict):
            return v[0]
    return {}

raw_df["details"] = raw_df["details"].apply(_parse_details)

# quick verification
print(type(raw_df["details"].iloc[0]))
raw_df["details"].head()
```

```python
import matplotlib.pyplot as plt

lengths = raw_df['combined_text'].astype(str).str.len()

fig, ax = plt.subplots(figsize=(10,4))
ax.hist(lengths, bins=50, color='C0', edgecolor='k')
ax.set_xlabel('Length of combined_text (characters)')
ax.set_ylabel('Count')
ax.set_title('Distribution of combined_text lengths')
plt.tight_layout()
plt.show()

print(lengths.describe())
```

```python
import re

def remove_surrogates(text):
    if isinstance(text, str):
        return re.sub(r"[\ud800-\udfff]", "", text)
    return text

raw_df = raw_df.applymap(remove_surrogates)
raw_df.to_csv("data/cleaned_product_data.csv", index=False)
```
